```python
"""
Text based adventure game: Ayesha, Zaynab, Mohammed & Tim
"""

from time import sleep

got_key = False  # Global variable to track key possession

def intro():
    print(r'''
*******************************************************************************
          |                   |                  |                     |
 _________|________________.=""_;=.______________|_____________________|_______
|                   |  ,-"_,=""     `"=.|                  |
|___________________|__"=._o`"-._        `"=.______________|___________________
          |                `"=._o`"=._      _`"=._                     |
 _________|_____________________:=._o "=._."_.-="'"=.__________________|_______
|                   |    __.--" , ; `"=._o." ,-"""-._ ".   |
|___________________|_._"  ,. .` ` `` ,  `"-._"-._   ". '__|___________________
          |           |o`"=._` , "` `; .". ,  "-._"-._; ;              |
 _________|___________| ;`-.o`"=._; ." ` '`."\` . "-._ /_______________|_______
|                   | |o;    `"-.o`"=._``  '` " ,__.--o;   |
|___________________|_| ;     (#) `-.o `"=.`_.--"_o.-; ;___|___________________
____/______/______/___|o;._    "      `".o|o_.--"    ;o;____/______/______/____
/______/______/______/_"=._o--._        ; | ;        ; ;/______/______/______/_
____/______/______/______/__"=._o--._   ;o|o;     _._;o;____/______/______/____
/______/______/______/______/____"=._o._; | ;_.--"o.--"_/______/______/______/_
____/______/______/______/______/_____"=.o|o_.--""___/______/______/______/____
/______/______/______/______/______/______/______/______/______/______/_____ /
*******************************************************************************
   _                                        _                  _
  | |                                      | |                | |
  | |_ _ __ ___  __ _ ___ _   _ _ __ ___   | |_   _   _ _ _   | |_  ___  _ __
  | __| '__/ _ \/ _` / __| | | | '__/ _ \  |  _ \| | | | '_ \ | __|/ _ \| '__/
  | |_| | |  __/ (_| \__ \ |_| | | |  __/  | | | | |_| | | | || |_|  __/| |
   \__|_|  \___|\__,_|___/\__,_|_|  \___|  |_| |_|\__,_|_| |_| \__|\___||_|

*******************************************************************************
      ''')
    print("""
    You are a treasure hunter on the search for the most valuable treasure known to man!
    Search the house for a key to unlock the chest of gold and jewels!
    But beware there are dangers lurking about...

    Press 'q' at any time to leave the game, Coward!
    """)
    input("Press any key to start your treasure hunt.")
    cellar()


def cellar():
    print("""
    Your search has led you to the fabled Fortnum Mansion.
    You are in the cellar. It's cold and damp and there is a strong earthy smell.
    It is so dark you can barely see your own hand, but you can make out some shapes
    from the moonlight shining through the broken window (your entry point).

    Make your move: (l)eft, (r)ight, (f)orward, (b)ackward
    """)
    
    while True:
        user_choice = input("Choose a direction (l/r/f/b): ").lower()

        if user_choice == 'q':
            quit_game()
            break

        elif user_choice in ['l', 'left', 'r', 'right', 'b', 'backward']:
            print("This way is blocked.")
        elif user_choice in ['f', 'forward']:
            print("\nYou enter the kitchen.")
            kitchen()
            break

        else:
            print("Invalid choice: (l)eft, (r)ight, (f)orward, (b)ackward")


def kitchen():
    print("""
    The kitchen smells of rot. Maggots squirm across the floor.
    A trap nearly caught your foot, and the floor feels dangerously weak.
    You see a sink, a fireplace, and a cupboard.
    """)
    
    while True:
        user_choice = input("Move on or explore: (s)ink, fire(p)lace, (c)upboard, (l/r/f/b)").lower()

        if user_choice == 'q':
            quit_game()
            break

        elif user_choice in ['s', 'sink']:
            print("The sink is full of dirty dishes.")
        elif user_choice in ['p', 'fireplace']:
            print("A rusty skeleton key hangs above the fireplace — purely ornamental.")
        elif user_choice in ['c', 'cupboard']:
            print("The cupboard contains delicate porceline.")

        elif user_choice in ['r', 'right']:
            print("This way is blocked.")
        elif user_choice in ['l', 'left']:
            print("\nYou enter the library.")
            library()
            break
        elif user_choice in ['b', 'backward']:
            print("\nYou return to the cellar.")
            cellar()
            break
        elif user_choice in ['f', 'forward']:
            print("\nYou enter the dining room.")
            dining()
            break

        else:
            print("Invalid choice: (s)ink, fire(p)lace, (c)upboard, (l/r/f/b)")


def library():
    global got_key
    print("""
    You are in the dusty library. A clean diary catches your attention —
    it's the mansion owner's, and he was a treasure hunter like you!
    There’s a desk, a chair, and a vase.
    """)

    while True:
        user_choice = input("Move on or explore: (d)esk, (c)hair, (v)ase, (l/r/f/b)").lower()

        if user_choice == 'q':
            quit_game()
            break

        elif user_choice in ['d', 'desk']:
            print("Inside the desk, you find a key. You take it.")
            got_key = True
        elif user_choice in ['c', 'chair']:
            print("The chair is surprisingly comfy.")
        elif user_choice in ['v', 'vase']:
            print("Fresh flowers. Someone’s been here recently...")
        
        elif user_choice in ['r', 'right']:
            print("\nYou return to the kitchen.")
            kitchen()
            break
        elif user_choice in ['f', 'forward']:
            print("\nYou enter the study.")
            study()
            break
        elif user_choice in ['l', 'left', 'b', 'backward']:
            print("This way is blocked.")

        else:
            print("Invalid choice: (d)esk, (c)hair, (v)ase, (l/r/f/b)")


def dining():
    print("""
    The dining room can host a banquet.
    A fireplace glows faintly. A portrait hangs nearby.
    """)

    while True:
        user_choice = input("Move on or explore: (fi)replace, (p)ortrait, (l/r/f/b)").lower()

        if user_choice == 'q':
            quit_game()
            break

        elif user_choice in ['fi', 'fireplace']:
            print("The fire is dwindling, but you appreciate it's warmth.")
        elif user_choice in ['p', 'portrait']:
            print("An elderly man holding a hunting rifle stares pompously from the frame.")

        elif user_choice in ['l', 'left', 'r', 'right', 'f', 'forward']:
            print("This way is blocked.")
        elif user_choice in ['b', 'backward']:
            print("\nYou reenter the kitchen.")
            kitchen()
            break
        
        else:
            print("Invalid choice: (fi)replace, (p)ortrait, (l/r/f/b)")


def study():
    print("""
    Cujo, the giant St Bernard, blocks your way! He growls deeply.
    A harp, a chain, and a club are nearby.
    """)

    while True:
        user_choice = input("Move on or act: (cl)ub, (ch)ain, (h)arp, (l/r/f/b)").lower()

        if user_choice == 'q':
            quit_game()
            break

        elif user_choice in ['cl', 'club', 'ch', 'chain', 'l', 'left', 'r', 'right', 'f', 'forward']:
            print("Cujo attacks you brutally...")
            sleep(3)
            game_over()
            break
        elif user_choice in ['h', 'harp']:
            print("The harp's tune soothes Cujo. He falls asleep.")
            input("Press Enter to tiptoe around him...")
            bedroom()
            break

        elif user_choice in ['b', 'back']:
            print("\nYou return to the library.")
            library()
            break

        else:
            print("Invalid choice: (cl)ub, (ch)ain, (h)arp, (l/r/f/b)")


def bedroom():
    global got_key
    print("You enter the bedroom. A large chest sits by the bed.")
    sleep(2)
    if got_key:
        win_game()
    else:
        print("It's locked and you don’t have the key.")
        sleep(2)
        print("Suddenly, Cujo appears. The last thing you see are his teeth...")
        game_over()


def win_game():
    print("You use the key to open the chest...")
    sleep(2)
    for _ in range(5):
        print(".", end=' ')
        sleep(1)
    print("\nYou've found the treasure! You Win!")


def game_over():
    print("""
    Cujo drags your lifeless body away.
    Game Over, brave (but dead) adventurer.
    """)


def quit_game():
    print("You flee the mansion. Maybe next time you'll be braver.")
    exit()


# Game loop
while True:
    intro()
    play_again = input("Play again? (p) or press any other key to exit: ").lower()
    got_key = False
    if play_again not in ['p', 'play']:
        print("Thanks for playing - Bye!")
        break
```
